## Corrélations

On calcule les corrélations linéaires des différentes colonnes avec la colonne à estimer : `ViolentCrimesPerPop`

```python
values = []
for column in communities_no_missing.columns.values:
    values.append(communities_no_missing['ViolentCrimesPerPop'].corr(communities_no_missing[column]))

correlation = pandas.DataFrame(values, 
                               index=communities_no_missing.columns.values, 
                               columns=['ViolentCrimesPerPop'])

correlation
```

Représentons les corrélations graphiquement :

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(10, 20))

sns.barplot(x='ViolentCrimesPerPop',
            y=correlation.index, 
            data=correlation, ax=ax)
plt.show()
```

```python
correl_trié = (correlation
                   .assign(abs=lambda df : abs(df['ViolentCrimesPerPop']))
                   .sort_values(by=['abs'], ascending=False)
                   .drop('abs', axis='columns')
                   [1:]
)
correl_trié
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
fig, ax = plt.subplots(figsize=(10, 20))

sns.barplot(x='ViolentCrimesPerPop',
            y=correl_trié.index, 
            data=correl_trié, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 20))

sns.barplot(x='ViolentCrimesPerPop',
            y=correl_trié.index, 
            data=correl_trié, ax=ax)

# labels
for i, (index, row) in enumerate(correl_trié.iterrows()):        
    ax.text(row['ViolentCrimesPerPop'], 
            i, 
            round(row['ViolentCrimesPerPop'], 2), 
    )

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 20))

sns.barplot(x='ViolentCrimesPerPop',
            y=correl_trié.index, 
            data=correl_trié, ax=ax)

# labels
for i, (index, row) in enumerate(correl_trié.iterrows()):
    if row['ViolentCrimesPerPop'] < 0:
        ha = 'right'
    else:
        ha = 'left'
        
    ax.text(row['ViolentCrimesPerPop'], 
            i, 
            round(row['ViolentCrimesPerPop'], 2), 
            horizontalalignment=ha,
            verticalalignment="center")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 20))

# figure
current_palette = sns.color_palette()
couleurs = []
for index, ligne in correl_trié.iterrows():
    couleurs.append(current_palette[int(10 * abs(ligne['ViolentCrimesPerPop']))])
sns.barplot(x='ViolentCrimesPerPop',
            y=correl_trié.index, 
            data=correl_trié, 
            palette=couleurs,
            ax=ax)

# labels
for i, (index, row) in enumerate(correl_trié.iterrows()):
    if row['ViolentCrimesPerPop'] < 0:
        ha = 'right'
    else:
        ha = 'left'
        
    ax.text(row['ViolentCrimesPerPop'], 
            i, 
            round(row['ViolentCrimesPerPop'], 2), 
            horizontalalignment=ha,
            verticalalignment="center")

plt.show()
```

La variable 'ViolentCrimesPerPop' est très corrélée négativement avec `PctKids2Par`et positivement avec `PctIlleg`. En revanche, elle n'est pas corrélée avec la variable `racePctAsian`.
